```python
#INDUSTRY
ind_df = pd.DataFrame(df.iloc[:, 3:19])
ind_df.columns
counts = ind_df.sum()
sorted_counts=counts.sort_values(ascending=True)
#sorted_counts.values
# Create a bar plot for the counts
colors="#d63627"
sns.set(style="ticks")  # Minimalist style without grid
font=20

plt.figure(figsize=(12,10))
bars = plt.barh(sorted_counts.index, sorted_counts.values, color=colors)
plt.ylabel('Industry')
plt.xlabel('Number of startups')
plt.xticks(rotation=90)
plt.yticks(fontsize=font)
plt.title("Count of Startups' Industries", fontsize=font)


for bar in bars:
    width = bar.get_width()
    plt.text(width + 1,  # Adjust to position text outside the bar
             bar.get_y() + bar.get_height() / 2,
             f'{int(width)}',  # Show as integer
             ha='left', va='center', fontsize=font - 6)

plt.tight_layout()
plt.show()
#plt.savefig("industry_chart.jpg", dpi=600, format='jpg', bbox_inches='tight')
```

In the German Tech Startup Ecosystem, `SaaS (Software as a Service)` is the most dominant industry with overwhelming startups operating in (315), almost double the second-popular industry - Smart Systems (177)

```python
#FUNDING STAGE
# Define the custom order for the 'Funding phase' categories
funding_order = ['Later Stage', 'Growth Stage','Early Stage','Seed','Pre-Seed' ]

# Group by 'Funding phase' and count the companies, without sorting by company count
funding = df.groupby('Funding phase').count()['Company']

# Reorder the funding phases according to the predefined order
funding = funding[funding_order]

# Create the horizontal bar plot
plt.figure(figsize=(12, 10))
bars = plt.barh(funding.index, funding.values, color=colors)

# Title and labels
plt.title("Funding phase of Startups", fontsize=font)
plt.xlabel('Number of startups', fontsize=font)
plt.ylabel('Funding phase', fontsize=font)
plt.yticks(fontsize=font)

# Add annotations to bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 2, bar.get_y() + bar.get_height() / 2, f'{int(width)}',
             ha='left', va='center', fontsize=font - 6)

# Display the plot
plt.tight_layout()
plt.savefig("Figure 1.5.jpg", dpi=600, format='jpg', bbox_inches='tight')
```

There is a large amount of young to extremely young businesses on the market, with 40% only in their `pre-seed` and `seed` funding stages.

The significant gap between 2 classes of the target variable (702:53) heightens the `need to handle heavy class imbalance`. 

```python
# Count occurrences of each value in 'percent_female'
percent_female = df['percent_female'].value_counts()

# Combine 0% into one category and all other values into another
percent_female_combined = percent_female.copy()
percent_female_combined.index = percent_female_combined.index.map(lambda x: 'No female founders' if x == 0 else 'There is at least one female founder')

# Plot the combined categories
plt.figure(figsize=(10, 10))
bars = plt.bar(percent_female_combined.index, percent_female_combined.values, color=colors)

# Title and labels
plt.title("Distribution of Percent Female in Startups", fontsize=font)
plt.xlabel('Appearance of female founders', fontsize=font)
plt.ylabel('Number of startups', fontsize=font)

# Add annotations to the bars
for bar in bars:
    height = bar.get_height()
    if bar.get_x() == bars[1].get_x():  # Check if it's the 'Non-0%' bar
        # Sum up the values of the 'Non-0%' category
        total_non_zero = percent_female_combined.drop('No female founders').sum()  # Sum all Non-0% values
        plt.text(bar.get_x() + bar.get_width() / 2, total_non_zero + 2, f'{int(total_non_zero)}',
                 ha='center', va='bottom', fontsize=font - 6)
    else:
        # Add individual annotations for '0%'
        plt.text(bar.get_x() + bar.get_width() / 2, height + 2, f'{int(height)}',
                 ha='center', va='bottom', fontsize=font - 6)

# Display the plot
plt.tight_layout()
plt.savefig("Figure 1.7.jpg", dpi=600, format='jpg', bbox_inches='tight')
```

The dataset also showed the `extreme disproportion of female leaders` in the high-tech domain, with only 74 among 755 firms (9.8%) having at least one female in their founder team.
